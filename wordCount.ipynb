```
fileLines.collect()
words = fileLines.flatMap(lambda fileln : fileln.split(" "))
words.count()
```

```
tuples = words.map(lambda wrd : (wrd,1))
```

```
cnts = tuples.reduceByKey(lambda a, b : a+b)
```

```
cnts.coalesce(1).saveAsTextFile("/FileStore/wrdCntOutputfinal")
```
